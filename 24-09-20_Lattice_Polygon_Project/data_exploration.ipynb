```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("data/lattice_shapes_data.xlsx")
df.insert(2, "Boundary Points",df["Vertex Points"] + df["Edge Points"])
df.insert(4, "Shape Points", 1/2*df["Boundary Points"] + df["Interior Points"] - 1)
df.drop(columns=["Vertex Points", "Edge Points"], inplace=True)
```

```python
sns.scatterplot(df, x="Shape Points", y="Area", hue="Interior Points")
plt.show()
# sns.scatterplot(df, x="Boundary Points", y="Shape Points", hue="Interior Points")
# plt.show()
```

```python
sns.lmplot(df, x="Boundary Plus Interior", y="Area")
```

```python
sns.pairplot(df, x_vars=df.columns[:-2], y_vars=["Area"])
```

```python
sns.pairplot(df[df.columns[:-1]])
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(df[["Boundary Points", "Interior Points"]], df["Area"])
df["Predicted Area"] = lr.predict(df[["Boundary Points", "Interior Points"]])
```
